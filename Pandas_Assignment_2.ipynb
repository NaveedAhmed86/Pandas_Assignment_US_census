```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
states0 = pd.read_csv('./states0.csv')
states1 = pd.read_csv('./states1.csv')
states2 = pd.read_csv('./states2.csv')
states3 = pd.read_csv('./states3.csv')
states4 = pd.read_csv('./states4.csv')
states5 = pd.read_csv('./states5.csv')
states6 = pd.read_csv('./states6.csv')
states7 = pd.read_csv('./states7.csv')
states8 = pd.read_csv('./states8.csv')
states9 = pd.read_csv('./states9.csv')
inventory = pd.read_csv('./inventory.csv')
```

```python
us_census = pd.concat([states0,states1,states2,states3,states4,states5,states6,states7,states8,states9])
us_census
```

```python
print(us_census.columns)
```

```python
us_census['Income'] = us_census.Income.str[1:]
us_census['Income'] = pd.to_numeric(us_census.Income)
us_census
```

```python
gender_split = us_census.GenderPop.str.split('_')
us_census['Men'] = gender_split.str.get(0)
us_census["Women"] = gender_split.str.get(1)
us_census
```

```python
us_census['Men'] = us_census.Men.str[:-1]
us_census['Women'] = us_census.Women.str[:-1]
us_census
```

```python
us_census['Men'] = pd.to_numeric(us_census.Men)
us_census['Women'] = pd.to_numeric(us_census.Women)
us_census
```

```python
us_census = us_census.fillna(value={'Women': us_census.TotalPop - us_census.Men})
us_census
```

```python
duplicates = us_census.duplicated(subset=['State'])
print(duplicates.value_counts())
us_census = us_census.drop_duplicates()
```

```python
plt.scatter(us_census['Women'], us_census['Income'],color=['red'])
plt.xlabel('Women')
plt.ylabel('Income')
plt.show()
```
